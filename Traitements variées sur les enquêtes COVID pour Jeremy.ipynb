# Fichier de traitements variés

```python
import pandas as pd
import pyshs
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf
import pyreadstat

def labels(x,meta):
    return meta.value_labels[meta.variable_to_label[x]]
```

## Enquête COCONEL juin

```python
data, meta = pyreadstat.read_sav("./Questionnaire chloro/Données/Data_CORO_V14.sav")

data["apportscience"] = data["OA195"].replace(labels("OA195"))
data["confiancescience"] = data["OA221"].replace(labels("OA221"))
```

Croisement apport science / confiance science

```python
pyshs.tableau_croise(data,"apportscience","confiancescience","RED")
```

## Enquêtes TRACTRUST

```python
data1, meta1 = pyreadstat.read_sav('./Questionnaire CS/conseilscientifique/SujetA.sav')
data2, meta2 = pyreadstat.read_sav('./Questionnaire CS/conseilscientifique/Data_118230.sav')

data1["religion"] = data1["REL1"].replace(labels("REL1",meta1))
data2["religion"] = data2["REL1"].replace(labels("REL1",meta2))

rec = {'Plutôt favorable':'2 - Plutôt favorable', 'Plutôt pas favorable':'3 - Plutôt pas favorable', 
 'Très favorable':'1 - Très favorable', 'Pas du tout favorable':'4 - Pas du tout favorable'}

data1["vac_general"] = data1['QA32'].replace(labels('QA32',meta1)).replace(rec)
data2["vac_general"] = data2['OA29'].replace(labels('OA29',meta2)).replace(rec)

# ATTENTION QUESTION PAS POSEE EN VAGUE 1
data1["vac_covid"] = data1['QA30'].replace({1.0: '1 - Oui, certainement',
                                             2.0: '2 - Oui, probablement',
                                             3.0: '3 - Non, probablement pas',
                                             4.0: '4 - Non, certainement pas'})
# 'QA31_r10': 'QA31 - Dans quelle mesure êtes-vous d’accord ou pas avec les propositions suivantes concernant ce futur vaccin et le COVID-19 ?  Une réponse par ligne. La vaccination contre le Covid-19 serait un bon moyen de protéger les personnes les plus fragiles',
data2["vac_covid"] = data2['OA28'].replace({1.0: '2 - Oui certainement',
                                             2.0: '3 - Oui, probablement',
                                             3.0: '4 - Non, probablement pas',
                                             4.0: '5 - Non, certainement pas',
                                             5.0: '1 - Vous êtes déjà vacciné (une dose ou plus)'})

reco = {1.0: '1 - Plus de bien que de mal',
 2.0: '3 - Plus de mal que de bien',
 3.0: '2 - A peu près autant de bien que de mal',
 4.0: '4 - Vous ne savez pas'}

data1["apportscience"] = data1["QA19"].replace(reco)
data2["apportscience"] = data2['OA20'].replace(reco)

reco = {1.0: '1 - Tout à fait confiance',
 2.0: '2 - Plutôt confiance',
 3.0: '3 - Plutôt pas confiance',
 4.0: '5 - Pas du tout confiance',
 5.0: '5 - Vous ne savez pas'}
data1["confiancescience"] = data1['QA9_r5'].replace(reco)
data2["confiancescience"] = data2['OA8_r5'].replace(reco)

data2["diplome"] = data2["DIPL2"].replace({1:"1 - inf bac",2:"1 - inf bac",3:"1 - inf bac",4:"1 - inf bac",
                                          5:"1 - inf bac", 6:"1 - inf bac",7:"1 - inf bac",8:"2 - bac",9:"2 - bac",
                                          10:"3 - sup bac",11:"3 - sup bac"})
data2["revenus"] = data2["REV"].replace({1.0: '2 - 0-1000€',
                                         2.0: '1 - 1000-2000€',
                                         3.0: '1 - 1000-2000€',
                                         4.0: '3 - 2000-4000€',
                                         5.0: '3 - 2000-4000€',
                                         6.0: '4 - 4000€ and more',
                                         7.0: '5 - NA'})
data2["csp"] = data2["PI1"].replace(labels("PI1",meta2))

```
